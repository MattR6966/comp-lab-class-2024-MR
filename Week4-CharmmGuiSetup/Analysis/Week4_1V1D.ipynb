```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Function to read .xvg files
def read_xvg(filename):
    data = []
    with open(filename, 'r') as f:
        for line in f:
            if line.startswith(("#", "@")):
                continue  # Skip comments and headers
            data.append([float(i) for i in line.split()])
    return np.array(data)

# Replace 'path/to/your/file.xvg' with the actual path to your .xvg file
xvg_file = '/home/mr6966/comp-lab-class/comp-lab-class-2024-MR/Week4-CharmmGuiSetup/1V1D_AMBER_GROMACS/gromacs/end2end.xvg'

# Read the data from the .xvg file
data = read_xvg(xvg_file)

# Assuming the first column is x and the second column is y
x = data[:, 0]
y = data[:, 1]

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(x, y, color='blue')
plt.xlabel('Time (ps)')  # Adjust label based on your data
plt.ylabel(r'C$\alpha$ Distance (nm)')       # Adjust label based on your data
plt.title('Distance between A1 and L31 Residues')
plt.grid()
plt.savefig('E2E.png')  # Save the plot as a PNG file
plt.show()
```
