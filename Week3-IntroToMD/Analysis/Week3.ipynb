```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Function to read .xvg files
def read_xvg(filename):
    data = []
    with open(filename, 'r') as f:
        for line in f:
            if line.startswith(("#", "@")):
                continue  # Skip comments and headers
            data.append([float(i) for i in line.split()])
    return np.array(data)

# Replace 'path/to/your/file.xvg' with the actual path to your .xvg file
xvg_file = '/home/mr6966/comp-lab-class/comp-lab-class-2024-MR/Week3-IntroToMD/Input/potential.xvg'

# Read the data from the .xvg file
data = read_xvg(xvg_file)

# Assuming the first column is x and the second column is y
x = data[:, 0]
y = data[:, 1]

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(x, y, label='Data from .xvg', color='blue')
plt.xlabel('Time (ps)')  # Adjust label based on your data
plt.ylabel('Energy(kJ/mol)')       # Adjust label based on your data
plt.title('TRP-Cage Energy Minimization')
plt.grid()
plt.savefig('Emin_plot.png')  # Save the plot as a PNG file
plt.show()
```
