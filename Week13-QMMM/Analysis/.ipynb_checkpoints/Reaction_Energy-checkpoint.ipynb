```python
import matplotlib.pyplot as plt
import numpy as np

# Load energy data (e.g., potential energy vs time)
npt_data = np.loadtxt('../Outputs/NPT-1.ener')
nvt_data = np.loadtxt('../Outputs/NVT-1.ener')
nvt_data[:, 1] /= 1000  
npt_data[:, 1] /= 1000  
# Plotting
plt.plot(nvt_data[:,1], nvt_data[:,4], label="NVT")
plt.plot(npt_data[:,1], npt_data[:,4], label="NPT")
plt.xlabel('Time (ps)')
plt.ylabel('Potential Energy (kcal/mol)')
plt.title('Potential Energy vs Time (ps)')
plt.legend()
plt.savefig('../Figures/MM_potential_energy_vs_time.png')
```
