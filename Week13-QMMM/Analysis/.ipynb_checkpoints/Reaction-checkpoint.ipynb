```python
import matplotlib.pyplot as plt
import numpy as np
# Load time and reaction coordinate data
reaction_data = np.loadtxt('../Outputs/monitor/MONITOR-COLVAR.metadynLog')
reaction_data[:, 0] /= 1000

# Plotting
plt.plot(reaction_data[:,0], reaction_data[:,1])
plt.xlabel('Time (ps)')
plt.ylabel('Bond Length Difference (Bohr)')
plt.savefig('../Figures/reaction_coordinate.png')
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Load time and reaction coordinate data
metadyn_data = np.loadtxt('../Outputs//metad1_initmonitor/METADYN-COLVAR.metadynLog')
freqhills_data = np.loadtxt('../Outputs//metad1_initmonitor_freqhills/METADYN-COLVAR.metadynLog')
reaction_data = np.loadtxt('../Outputs/monitor/MONITOR-COLVAR.metadynLog')
freqhills_data[:, 0] /= 1000
reaction_data[:, 0] /= 1000
metadyn_data[:, 0] /= 1000
# Plotting
plt.plot(metadyn_data[:,0], metadyn_data[:,1], label = 'QM/MM')
plt.plot(freqhills_data[:,0], freqhills_data[:,1], label = 'Metadynamics freq = 20')
plt.plot(reaction_data[:,0], reaction_data[:,1], label = 'Metadynamics freq = 200')
plt.xlabel('Time (ps)')
plt.legend()
plt.ylabel('Bond Length Difference (Bohr)')
plt.savefig('../Figures/reaction_coordinate_comparison.png')

```
