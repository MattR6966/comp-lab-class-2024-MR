```python
import mdtraj as md
import matplotlib.pyplot as plt
# Load beta amyloid trajectory (assuming it's in 'beta_amyloid.xtc' and structure '1hz3.pdb')
amyloid_traj = md.load('/scratch/work/courses/CHEM-GA-2671-2024fa/students/mr6966/comp-lab-class-2024-MR/Week2-DataAnalysis/Data/1hz3_T310.stepid25000000.every100ps.nowater.xtc', top='/scratch/work/courses/CHEM-GA-2671-2024fa/students/mr6966/comp-lab-class-2024-MR/Week2-DataAnalysis/Data/1hz3_T310.start.nowater.gro')

# Compute the end-to-end distance
end_to_end = md.compute_distances(amyloid_traj, [[0, 17]])  # Assuming atoms 0 and -1 are the ends

# Compute the radius of gyration
radius_of_gyration = md.compute_rg(amyloid_traj)

# Plot end-to-end distance and radius of gyration vs time
plt.figure(figsize=(8, 6))
plt.plot(amyloid_traj.time, end_to_end, label='End-to-End Distance')
plt.plot(amyloid_traj.time, radius_of_gyration, label='Radius of Gyration')
plt.xlabel('Time (ps)')
plt.ylabel('Distance / Radius (nm)')
plt.title('End-to-End Distance and Radius of Gyration vs Time')
plt.legend()
plt.grid(True)
plt.show()

# Normalize and plot histograms
plt.figure(figsize=(8, 6))
plt.subplot(1, 2, 1)
plt.hist(end_to_end.flatten(), bins=30, density=True)
plt.title('Normalized Histogram of End-to-End Distance')

plt.subplot(1, 2, 2)
plt.hist(radius_of_gyration.flatten(), bins=30, density=True)
plt.title('Normalized Histogram of Radius of Gyration')

plt.tight_layout()
plt.show()
```
